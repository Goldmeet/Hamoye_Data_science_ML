**GROUPING OF DATA**

```python
GROUP_AREA = df.groupby(['Area', 'Y2017'])
```

```python
GROUP_AREA.first()
```

```python
AREA = df.groupby(['Area']).agg({'Y2017': 'sum'})
```

```python
AREA
```

```python
Grouped_Element = df.groupby(['Element', 'Area']).agg({'Y2015': 'sum'})
```

**TOTAL STOCK VARIATIOIN ACROSS THE YEARS**
